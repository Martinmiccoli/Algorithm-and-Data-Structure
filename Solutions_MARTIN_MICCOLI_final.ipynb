### Exercise: Selection Sort
Write the function ```SelectionSort(coll)``` that returns a sorted list with the elements in *coll*. 
You have to implements Selection Sort algorithm.

`````python
def SelectionSort(my_list):
    for elemento_corrente in range(len(my_list)):
        posizione_del_minimo = elemento_corrente                    
        for elemento_corrente_successivo in range(elemento_corrente+1, len(my_list)): 
            if my_list[elemento_corrente_successivo] < my_list[posizione_del_minimo]:   
                posizione_del_minimo = elemento_corrente_successivo 
        my_list[posizione_del_minimo], my_list[elemento_corrente] = my_list[elemento_corrente], my_list[posizione_del_minimo]
    return my_list 

#====================================================================


def test_sortedness(my_list):
    return my_list == sorted(my_list)
my_list = list(range(10))[::-1]
print(SelectionSort(my_list))
assert test_sortedness( SelectionSort(my_list) ), "Must be increasing!"


assert SelectionSort([]) == [] 
assert SelectionSort([1]) == [1] 
assert SelectionSort([1, 2]) == [1, 2] 
assert SelectionSort([2, 1]) == [1, 2] 
assert SelectionSort([2, 2, 2, 2]) == [2, 2, 2, 2] 
assert SelectionSort([-4, -3, -2, -1, 0, 1, 2, 3, 4]) == [-4, -3, -2, -1, 0, 1, 2, 3, 4]
assert SelectionSort([4, 3, 2, 1]) == [1, 2, 3, 4] 
assert SelectionSort([3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5]) == [1, 1, 2, 3, 3, 4, 5, 5, 5, 6, 9] 
assert SelectionSort(list(range(1000, 0, -1))) == list(range(1, 1001)) 
`````

### Exercise: Insertion Sort
Write the function ```InsertionSort(coll)``` that returns a sorted list with the elements in *coll*. 
You have to implements Insertion Sort algorithm.

`````python
def InsertionSort(my_list):
    for indice_corrente in range(1, len(my_list)):
        elemento_corrente = my_list[indice_corrente]
        indice_precedente = indice_corrente - 1
        while indice_precedente > -1 and my_list[indice_precedente] > elemento_corrente: 
            my_list[indice_precedente + 1] = my_list[indice_precedente]  
            indice_precedente -= 1  
        my_list[indice_precedente + 1] = elemento_corrente 
    return my_list

#====================================================================


my_list = list(range(10))[::-1]

print(InsertionSort(my_list))

assert test_sortedness( InsertionSort(my_list) ), "Must be increasing!"


assert InsertionSort([]) == [] 
assert InsertionSort([1]) == [1]
assert InsertionSort([1, 2]) == [1, 2] 
assert InsertionSort([2, 1]) == [1, 2] 
assert InsertionSort([2, 2, 2, 2]) == [2, 2, 2, 2] 
assert InsertionSort([-4, -3, -2, -1, 0, 1, 2, 3, 4]) == [-4, -3, -2, -1, 0, 1, 2, 3, 4] 
assert InsertionSort([4, 3, 2, 1]) == [1, 2, 3, 4] 
assert InsertionSort([3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5]) == [1, 1, 2, 3, 3, 4, 5, 5, 5, 6, 9] 
assert InsertionSort(list(range(1000, 0, -1))) == list(range(1, 1001)) 
`````

`````python
def fractional_knapsack(L, W):
    L.sort(key=lambda x: x[0] / x[1], reverse=True)  
    tot_valore = 0 
    capacita_rimanente = W 

    for valore, peso in L:
        if capacita_rimanente >= peso: 
            tot_valore += valore 
            capacita_rimanente -= peso 
        else:
            frazione_ultimo_elemento = capacita_rimanente / peso
            tot_valore += valore * frazione_ultimo_elemento 
            break 
    return tot_valore 

#====================================================================

L = [(60, 10), (100, 20), (120, 30)]
assert fractional_knapsack(L, 50) == 240.0, "Fail!"
L = [(30, 5), (40, 10), (45, 15), (77, 22), (90, 25)]
assert fractional_knapsack(L, 60) == 230.0, "Fail!"
assert fractional_knapsack(L, 15) == 70.0,  "Fail!"
assert fractional_knapsack(L, 10) == 50.0,  "Fail!"

# Test case where all items have the same value-to-weight ratio
L = [(10, 5), (20, 10), (30, 15)]
assert fractional_knapsack(L, 50) == 60.0, "Fail! All items have the same value-to-weight ratio"

# Test case where the capacity is larger than the sum of all weights
L = [(10, 5), (20, 10), (30, 15)]
assert fractional_knapsack(L, 100) == 60.0, "Fail! Capacity larger than sum of all weights test case"
`````

### Exercise: Pareto frontier of a set of points in 2-D space (aka Skyline problem)
We are given a set $S$ of $n$ 2D points.
A point $(x,y)$ dominates a point $(x',y')$ iff $ùë•'\leq ùë•$ and $y'\leq ùë¶$. 
Our goal is to find the set $P$ of dominating points in $S$. 
This corresponds to find the Pareto frontier (or, equivalently, the skyline). 

The problem can be solved in $\Theta(n\log n)$ time.

To find $P$ we need to sort points in $S$ by $x$ in descending order, 
and if $x$‚Ä≤ùë† the same by $y$ in descending order. This takes $\Theta(n\log n)$ time. 
Then, we do the following.

- Include first point in $P$ and remember this point as $ùëá$. 
- Iterates through the point (let $C$ current point):
* if $C$ is dominated by $T$, then skip $C$ and go to next point;
* Otherwise, include $C$ in $P$ and set $ùëá=ùê∂$.

This step can be performed in linear time.

Implement the function ```pareto_frontier(S)```, which returns the pareto frontier $P$ of the points in $S$.


`````python
def pareto_frontier(S):
    S = sorted(S, key=lambda x: (x[0], x[1]), reverse=True) 
    if S == []:
        return []
    P = [S[0]]
    T = S[0]
    for C in S[1:]:
        if C[0] <= T[0] and C[1] <= T[1]:
            continue
        P.append(C)
        T = C

    P.sort()
    return P

#====================================================================


S = [(6, 7.5), (7, 8), (8, 7), (2, 9), (3, 9.5), (1, 10), (4, 9), (5, 8)]
assert pareto_frontier(S) == [(1, 10), (3, 9.5), (4, 9), (7, 8), (8, 7)], "Fail!"


assert pareto_frontier([]) == [], "Fail! Empty list test failed."
assert pareto_frontier([(1, 1)]) == [(1, 1)], "Fail! Single point test failed."
assert pareto_frontier([(1, 5), (2, 5), (3, 5), (4, 5)]) == [(4, 5)], "Fail! Same y-value test failed."
assert pareto_frontier([(5, 1), (5, 2), (5, 3), (5, 4)]) == [(5, 4)], "Fail! Same x-value test failed."
assert pareto_frontier([(1, 6), (2, 5), (3, 4), (4, 3), (5, 2), (6, 1)]) == [(1, 6), (2, 5), (3, 4), (4, 3), (5, 2), (6, 1)], "Fail! All Pareto points test failed."
assert pareto_frontier([(6, 7.5), (7, 8), (8, 7), (2, 9), (3, 9.5), (1, 10), (4, 9), (5, 8)]) == [(1, 10), (3, 9.5), (4, 9), (7, 8), (8, 7)], "Fail! Random points test failed."
`````
